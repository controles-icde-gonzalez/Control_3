```python
# 1. Integrar archivos 
#instalar bibliotecas para manipular el documento  en excel y tablas de datos
!pip install pandas openpyxl
import pandas as pd
level1 = pd.read_excel('level1.xlsx', header = [4]) # very high human development
level2 = pd.read_excel('level2.xlsx', ) # high human development
level3 = pd.read_excel('level3.xlsx') # medium human development
level4 = pd.read_excel('level4.xlsx') # low human development
```

```python
# Para level1 quitar esas filas que eran del encabezado pero ahora estan como datos de la tabla
level1 = level1.drop(level1.index[[0, 1]])
level1.reset_index(drop=True, inplace=True)
level1
```

```python
# Para level2: quitar dos primeras filas tambien, no hay cambios por hacer en el encabezado
level2 = level2.drop(level2.index[[0, 1]])
level2.reset_index(drop=True, inplace=True)
level2
```

```python
# Para level3: igualmente, solo quitar las dos primeras filas
level3 = level3.drop(level3.index[[0, 1]])
level3.reset_index(drop=True, inplace=True)
level3
```

```python
# Para level4: No hay filas por quitar
level4
```

```python
# PARA INTEGRAR LOS DATA FRAME: solo con las columnas en comun
# interseccion de columnas con el mismo nombre
columnas_comun = list(set(level1.columns) & set(level2.columns) & set(level3.columns) & set(level4.columns))
columnas_comun
# cambio para solo quedarme con esas columnas
level1 = level1[columnas_comun]
level2 = level2[columnas_comun]
level3 = level3[columnas_comun]
level4 = level4[columnas_comun]

# integrar :)
integrar_levels = pd.concat([level1, level2, level3, level4], axis=0)
integrar_levels

#cambiar nombres para simplificar 
integrar_levels = integrar_levels.rename(columns={'HDI rank': 'HDI_Rank_2021',
                                                  'Gross national income (GNI) per capita': 'GNI_2021',
                                                  'Life expectancy at birth': 'LEB_2021',
                                                  'Mean years of schooling' : 'MYS_2021',
                                                  'Human Development Index (HDI) ': 'HDI_2021',
                                                  'Expected years of schooling': 'EYS_2021',
                                                  'GNI per capita rank minus HDI rank' : 'GNI-HDIRank'})
#cambiar el orden de las columnas
nuevo_orden = ['HDI_Rank_2021', 'Country', 'HDI_2021', 'LEB_2021', 'EYS_2021', 'MYS_2021', 'GNI_2021', 'GNI-HDIRank']

integrar_levels = integrar_levels[nuevo_orden]

# cambiar los nombres de los paises a mayusculas
integrar_levels['Country'] = integrar_levels['Country'].str.upper()

integrar_levels
```

```python
# 2. Merge con countrycodes
countryCodes = pd.read_excel('countryCODES.xlsx')

countryCodes
```

```python
# Cambio de nombre a paises en integrar_levels a paises que pueden causar problemas
troublesome={'HONG KONG, CHINA (SAR)':'CHINA','TÃœRKIYE':'TURKEY', ' NAMIBIA ' : 'NAMIBIA'}

integrar_levels = integrar_levels.replace(to_replace={'Country':troublesome})
```

```python
!pip install thefuzz
from thefuzz import process as fz
import pandas as pd

# IteraciÃ³n para que el diccionario country_mapping se llene a partir de las similitudes entre las columnas country de integrar_levels y countryCodes
country_mapping = {}
for country in countryCodes['Country'].unique():
    match = fz.extractOne(country, integrar_levels['Country'].unique())
    if match[1] > 90:
        country_mapping[country] = match[0]

# Uniformizar nombres de paises para despues realizar merge
# cambia los nombres de paises de countrycodes a los que encontre con similitud>80 con integrar_levels
countryCodes['Country'] = countryCodes['Country'].map(country_mapping).fillna(countryCodes['Country'])

# merge
merged_df = pd.merge(integrar_levels, countryCodes, on='Country', how='left')

# quitar las columnas con missing values que por defecto dejo el diccionario country_mapping
merged_df = merged_df.dropna(subset=['Officialstatename'])

merged_df
```

```python
# 3. Para hacer reporte mensual con fallecidos
fallecidos_covid = pd.read_csv('fallecidosCOVID.csv', delimiter=';')

# como necesito separar la fecha, usando formateo de fechas
fallecidos_covid['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos_covid['FECHA_FALLECIMIENTO'], format='%Y%m%d') 

fallecidos_covid['AÃ±o'] = fallecidos_covid['FECHA_FALLECIMIENTO'].dt.year
fallecidos_covid['Mes'] = fallecidos_covid['FECHA_FALLECIMIENTO'].dt.month
fallecidos_covid['DÃ­a'] = fallecidos_covid['FECHA_FALLECIMIENTO'].dt.day

# aÃ±adir columna con nombres de meses
fallecidos_covid['Nombre_Mes'] = fallecidos_covid['Mes'].apply(lambda x: {
    1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril', 5: 'Mayo', 6: 'Junio',
    7: 'Julio', 8: 'Agosto', 9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'}[x])

fallecidos_covid
```

```python
# para crear el reporte mensual con reshaping - tablas pivot
reporte_3 = fallecidos_covid.pivot_table(
    index=['Mes', 'Nombre_Mes', 'AÃ±o'],  # indice por aÃ±o y mes
    aggfunc='size',           # contar los fallecidos
    fill_value=0              # rellenar con 0 si hay datos faltantes
).reset_index(name='cant_fallecidos')

# para que muestre los datos de manera ordenada de meses y aÃ±os
reporte_3 = reporte_3.sort_values(by=['AÃ±o', 'Mes'], ascending=True)

# resetear el index
reporte_3 = reporte_3.reset_index(drop=True)

reporte_3
```

```python
# 4.  Reporte de fallecidos por mes, pero el tipo de muerte debe estar en columnas separadas

# Re-shaping - tabla pivot
reporte_4 = fallecidos_covid.pivot_table(
    index=['Mes', 'Nombre_Mes', 'AÃ±o'],              # Ãndice por AÃ±o y Mes
    columns='CLASIFICACION_DEF',          # Columnas por tipo de muerte
    aggfunc='size',                       # Contar los fallecidos
    fill_value=0                          # Rellenar con 0 si no hay datos
).reset_index()

# para que muestre los datos de manera ordenada de meses y aÃ±os
reporte_4 = reporte_4.sort_values(by=['AÃ±o', 'Mes'], ascending=True)
reporte_4 = reporte_4.reset_index(drop=True)

reporte_4.columns.name = None

# cambiar nombres
reporte_4 = reporte_4.rename(columns={'Criterio SINADEF': 'C_SINADEF',
                                    'Criterio clÃ­nico': 'C_Clinico',
                                    'Criterio investigaciÃ³n EpidemiolÃ³gica': 'C_IEpidemiologica',
                                    'Criterio nexo epidemiolÃ³gico' : 'C_NEpidemiologico',
                                    'Criterio radiolÃ³gico': 'C_Radiologico',
                                    'Criterio serolÃ³gico': 'C_Serologico',
                                    'Criterio virolÃ³gico' : 'C_Virologico',
                                     'sinadef': 'Sinadef'})


reporte_4
```
